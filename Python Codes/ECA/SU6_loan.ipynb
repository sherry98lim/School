```python
from sklearn import tree
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('loan.csv')
df
```

```python
# Extract the dependent and independent variables
X_origin = df.drop('Loan Approved', axis=1)
y = df['Loan Approved']
```

```python
#data preparation -> get dummy variables
X = pd.get_dummies(X_origin)
X
```

```python
# Create a decision tree classifier
dtc = tree.DecisionTreeClassifier(random_state=123)
```

```python
# Fit the decision tree
dtc.fit(X, y)
```

```python
# from sklearn import plot_tree
# control the plotting size 
plt.figure(figsize=(10, 8))
# Plot the decision tree
tree.plot_tree(dtc, filled=True, feature_names=list(X.columns), class_names=list(y.unique()))#,max_depth=2
```

```python
# from sklearn import plot_tree
# control the plotting size 
plt.figure(figsize=(10, 8))
# Plot the decision tree
tree.plot_tree(dtc, filled=True, feature_names=list(X.columns), class_names=list(y.unique()), max_depth=2)
```
